```python
# storing trace
import json
from seal.task_result import TaskResult
from seal.database import init_db, insert_task_result

init_db()
results = []

jsonl_path = "/content/baseline_all_traces.jsonl"
with open(jsonl_path) as f:
    for line in f:
        d = json.loads(line)
        result = TaskResult.from_dict(d)
        insert_task_result(result)
        results.append(result)

print(f"Loaded and stored {len(results)} tasks.")
```

```python
from seal.metrics import task_success_rate, failure_analysis_precision, judge_alignment, convergence_speed

print("Success rate:", task_success_rate(results))
print("Failure precision:", failure_analysis_precision(results))
print("Judge alignment:", judge_alignment(results))
print("Convergence:", convergence_speed(results))
```
